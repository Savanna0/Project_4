```python
import pandas as pd
from pathlib import Path
```

```python
# Store filepath in a variable
housing_csv_path = Path("Resources/RDC_Inventory_Core_Metrics_County_History.csv")
```

```python
# Read our data file with the Pandas library
df = pd.read_csv(housing_csv_path)
```

```python
# Show the first five rows.
df.head()
```

```python
df.info()
```

```python
#Dropping columns that have empty cells over the 50% threshold.
threshold = 0.5
columns_to_drop = df.columns[df.isnull().mean() > threshold]
data_cleaned = df.drop(columns=columns_to_drop)
data_cleaned.head()
```

```python
removed_columns = list(columns_to_drop)
removed_columns
```

```python
#only keeping rows where quality flag equals 0. If there was a quality flag it would have been an outlier where the home was overpriced or under priced. 
df_test = df[df['quality_flag'] != 1]
df_test.head(10)
```

```python
df_test.info()
```

```python
#dropping columns that are not needed
cleaned_df= df_test.drop(['median_listing_price', 'median_listing_price_mm','median_listing_price_yy','new_listing_count_mm',
              'new_listing_count_yy','pending_ratio','pending_ratio_mm','pending_ratio_yy','pending_listing_count','pending_listing_count_mm',
              'pending_listing_count_yy','quality_flag','median_listing_price_mm','median_listing_price_yy','active_listing_count_mm','active_listing_count_yy',
              'median_days_on_market_mm','median_days_on_market_yy','new_listing_count_mm','new_listing_count_yy','price_increased_count_mm',
              'price_increased_count','price_increased_count_yy','price_reduced_count','price_reduced_count_mm','price_reduced_count_mm',
              'price_reduced_count_yy','median_listing_price_per_square_foot_mm','median_listing_price_per_square_foot_yy','median_square_feet_mm',
              'median_square_feet_yy','average_listing_price_mm','average_listing_price_yy','total_listing_count_mm','total_listing_count_yy',
              'pending_ratio_mm','pending_ratio_yy','county_fips'], axis=1)
```

```python
cleaned_df.info()
```

```python
cleaned_df
```

```python
#Dropping rows with NaN values
df_cleaned = cleaned_df.dropna()
df_cleaned.head()
```

```python
#Renaming columns for readability.
df_renamed = df_cleaned.rename(columns={"month_date_yyyymm": "Date YYYYMM","county_fips": "County FIPS", "county_name":"County Name","active_listing_count":"Active Listing Count",
                                "active_listing_count_mm":"Active Listing Count M/M","active_listing_count_yy":"Active Listing Count Y/Y",
                                "median_days_on_market":"Median Days on Market","median_days_on_market_mm":"Median Days on Market M/M","median_days_on_market_yy":"Median Days on Market Y/Y",
                                "new_listing_count":"New Listing Count","price_increased_count":"Price Increase Count","price_increased_count_mm":"Price Increase Count M/M",
                                "price_increased_count_yy":"Price Increase Count Y/Y","price_reduced_count":"Price Decrease Count","price_reduced_count_mm":"Price Decrease Count M/M",
                                "price_reduced_count_yy":"Price Decrease Count Y/Y","median_listing_price_per_square_foot":"Median List Price Per Sqft",
                                "median_listing_price_per_square_foot_mm":"Median List Price Per Sqft M/M","median_listing_price_per_square_foot_yy":"Median List Price Per Sqft Y/Y",
                                "median_square_feet":"Median Sqft","median_square_feet_mm":"Median sqft M/M","median_square_feet_yy":"Median Sqft Y/Y",
                                "average_listing_price":"Avg Listing Price","average_listing_price_mm":"Avg Listing Price M/M","average_listing_price_yy":"Avg Listing Price Y/Y",
                                "total_listing_count":"Total Listing Count","total_listing_count_mm":"Total Listing Count M/M","total_listing_count_yy":"Total Listing Count Y/Y"})
```

```python
df_renamed.head()
```

```python
#df_renamed['Year'] = df_renamed['Date YYYYMM'].str[:4].astype(int)
#df_renamed['Month'] = df_renamed['Date YYYYMM'].str[4:].astype(int)

df_renamed['Year'] = df_renamed['Date YYYYMM'].astype(str).str[:4].astype(int)  # Convert the year to integer
df_renamed['Month'] = df_renamed['Date YYYYMM'].astype(str).str[4:6].astype(int)  # Convert the month to integer
```

```python
df_renamed
```
